# Non-Linear Activation Functions

I'll focus on **graphing two non-linear activation functions**. The two most commonly used activation functions are the **Sigmoid** and the **Rectified Linear Unit (ReLU)**.

The **Sigmoid** function maps input values to an output between 0 and 1, making it useful in models where we need probabilities. It is defined as $\sigma(x) = \frac{1}{1 + e^{-x}}$, where large positive inputs map to 1 and large negative inputs to 0. This characteristic can lead to the **vanishing gradient problem**, making it less effective for deep networks.

The **ReLU** activation function, on the other hand, is defined as $\text{ReLU}(x) = \max(0, x)$. It has the advantage of being computationally efficient and mitigating the vanishing gradient issue, though it can suffer from the **dying ReLU problem** when neurons get stuck in the zero state.

By graphing both functions over a range of values, we can visually compare their behaviors and understand their impact on neural network training.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the input range for plotting
x = np.linspace(-10, 10, 400)

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU Activation Function
def relu(x):
    return np.maximum(0, x)

# Derivatives of the activation functions (for deeper insight into gradients)
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Create a plot with subplots to visualize both functions and their gradients
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot Sigmoid function
axs[0, 0].plot(x, sigmoid(x), label="Sigmoid", color='b')
axs[0, 0].set_title('Sigmoid Activation Function')
axs[0, 0].set_xlabel('Input (x)')
axs[0, 0].set_ylabel('Output')
axs[0, 0].grid(True)
axs[0, 0].legend()

# Plot ReLU function
axs[0, 1].plot(x, relu(x), label="ReLU", color='r')
axs[0, 1].set_title('ReLU Activation Function')
axs[0, 1].set_xlabel('Input (x)')
axs[0, 1].set_ylabel('Output')
axs[0, 1].grid(True)
axs[0, 1].legend()

# Plot the derivative of Sigmoid
axs[1, 0].plot(x, sigmoid_derivative(x), label="Sigmoid Derivative", color='b')
axs[1, 0].set_title('Sigmoid Derivative')
axs[1, 0].set_xlabel('Input (x)')
axs[1, 0].set_ylabel('Gradient')
axs[1, 0].grid(True)
axs[1, 0].legend()

# Plot the derivative of ReLU
axs[1, 1].plot(x, relu_derivative(x), label="ReLU Derivative", color='r')
axs[1, 1].set_title('ReLU Derivative')
axs[1, 1].set_xlabel('Input (x)')
axs[1, 1].set_ylabel('Gradient')
axs[1, 1].grid(True)
axs[1, 1].legend()

# Adjust layout for better visibility
plt.tight_layout()
plt.show()
```

The plot contains four subplots:

**1. Sigmoid Activation Function**: Shows the S-shaped curve of Sigmoid.

**2. ReLU Activation Function**: Displays the sharp turn at 0 of the ReLU function.

**3. Sigmoid Derivative**: Highlights how the gradient shrinks for large or small inputs.

**4. ReLU Derivative**: Displays the flat gradient of ReLU for negative inputs and a constant gradient of 1 for positive inputs.

* The **Sigmoid** function is often used in **binary classification** problems. Its derivative, however, tends to vanish for large positive or negative inputs, which is known as the **vanishing gradient problem**. This behavior is critical in training deep networks, as the gradients become so small that weight updates become ineffective (as mentioned in the Dive into Deep Learning).

* The **ReLU** function, in contrast, doesnt suffer from vanishing gradients for positive values, allowing for faster training. However, its **zero gradient for negative values** can cause the dying ReLU problem. This is why variants like **Leaky ReLU** or **Parametric ReLU** are sometimes used to mitigate this issue.
